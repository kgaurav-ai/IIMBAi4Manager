```python
imdb_df['genres_cnt'] = imdb_df['genres'].str.count('\|')
imdb_df.head(5)
imdb_df['genres_cnt'].describe()
```

```python
imdb_df[['genres','duration']]
```

```python
imdb_df['genre_list'] = imdb_df['genres'].str.split('\|')
imdb_df['genre_list'].head()
```

```python
imdb_df_exploded = imdb_df.explode('genre_list')
imdb_df_exploded
```

```python
genre_status = imdb_df_exploded.groupby('genre_list')['duration'].agg(['mean', 'median', 'count','std','min','max'])
genre_status.sort_values(by='mean')
```
